# Verify Central Limit Theorem

CLT claims that if a variable $X$ is the sum of **$N$** random numbers that are indepenendently and identically drawn from a probability distribution, then $X$ follows a Gaussian distribution when $N$ is large

$$X_N = \frac{1}{N}\sum_{i=0}^{N} x_i$$

where $ x_i $ is drawn from $ f(x)$ which has a mean of $\mu$ and a std. dev of $\sigma$

The distribution of $X_N$ should have a mean of $\mu$ and a std. dev. of $\sigma/\sqrt{N}$


In this exercise, we generate $x_i$ from a uniform distribution between 0 and 1, and we compare the distribution of $X_N$ with a Gaussian distribution with its mean and sigma predicted by the Central Limit Theorem.

### Cumulative distribution function (CDF)

We develop code to calculate CDF here.

Recall the definition of the CDF for a given probability distribution $f(x)$
$$F(a) = \int_{-\infty}^{a} f(x)dx$$

First, we want to calculate the CDF "manualy" for a Poisson distribution with a mean of 100. Then, we will use the cdf method of scipy.stats to get the CDF values. The manual calculation serves as a cross check.

```python
# what is the probability when expectation is 100, and the observation is at most 90?
# in other words, integrate the distribution Poisson (x, 100) from 0 to 90 (including 90)

# Generate a sequence of observed values between 0 and 150. They are all integers 
#Fix this line
obs = np.linspace()

# Calculate the poisson probability values at these 151 points
# the second input argument is the mean value, which should be 100
prob = poisson.pmf(obs,100)

# Combine these two arrays into one with a shape of (151,2)
Prob_vs_Obs = np.vstack((obs,prob)).T

# Develop the code below
# Sum the probabilities values between 0 and 90

# Fix this line
prob_atmost90=np.sum(prob[])
print('Probability is %.4f '% prob_atmost90)
```

### Pseudo experiment

If the outcome of an experiment can be modeled by a probability function, then we can use the probability function to generate random outcomes that represent possible outcomes of an experiment. These random outcomes are called `pseudo experiments` or `toy Monte Carlo`.

In this exercise, our experiment measures a quantity that follows a Poisson distribution with a mean of 100. Each sample consists of 5000 outcomes of this measurement. For each sample, we determine the fraction of outcomes that are less than or equal to 90.

```python
# assume our experiment has an expectation of 100 
# the outcome follows a Poisson distribution 
# now let's generate random outcomes from a Poisson function with mean of 100

# Create a random number generator with a seed of 12
rng = np.random.default_rng(seed=12)
```

```python
# check out the shape and size of these two arrays sampleP1 and sampleP2
sampleP1=rng.poisson(100,size=5000)
sampleP2=rng.poisson(100,size=(2,5000))

```
